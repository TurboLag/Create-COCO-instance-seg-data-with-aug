```python
def map_to_bound5(array1):
    
    image_draw = Image.new("1",(1024,1024), "#000000")  
    draw = ImageDraw.Draw(image_draw)
    draw.polygon((array1[1].tolist()),  outline= 'white' , fill= 'white' )

    return np.asarray(image_draw).astype(np.uint8)
```

```python
image_id = 1
segmentation_id = 1


# go through each image
for image_filename in progress_bar(train_image):
    image = Image.fromarray(image_filename)
    image_info = pycococreatortools.create_image_info(
        image_id, str(image_id) + '.tif' , image.size)
    coco_output["images"].append(image_info)

    for anno_info in train_txt[image_id-1]:

        category_info = {'id':  float(anno_info[0]), 'is_crowd': False }

        binary_mask = map_to_bound5(anno_info)

        annotation_info = pycococreatortools.create_annotation_info(
            segmentation_id, image_id, category_info, binary_mask,
            image.size, tolerance=2)

        if annotation_info is not None:
            coco_output["annotations"].append(annotation_info)

        segmentation_id = segmentation_id + 1

    image_id = image_id + 1

```
