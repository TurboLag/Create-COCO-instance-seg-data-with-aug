```python
def color(i):
    if(i == 1):
        outputcolor = '#FF0000'
    if(i == 2):
        outputcolor = '#00FF00'
    if(i == 3):
        outputcolor = '#0000FF'
    if(i == 4):
        outputcolor = '#7F0000'
    if(i == 5):
        outputcolor = '#00007F'
    return outputcolor

def map_to_bound5(image, array1):
    colors = (color(array1[0]))
#     image_draw = Image.new("L",(1024,1024,3), "#000000")  
    image_draw = Image.fromarray(image)
    draw = ImageDraw.Draw(image_draw)
#     print((array[1:9]))
    draw.polygon((array1[1:9].tolist()),  outline= colors , fill= colors )
    # image_draw.show()
    arrayreturn = copy.deepcopy(np.asarray(image_draw))
#     arrayreturn = arrayreturn//255
    return arrayreturn

def map_to_bound(array1):
    image_draw = Image.new("L",(1024,1024), "#000000")  

    draw = ImageDraw.Draw(image_draw)

    draw.polygon((array1[1:9].tolist()),  outline= "white" , fill= 'white' )

    return np.asarray(image_draw)//255

def map_to_bound(array1):
    image_draw = Image.fromarray(np.zeros((1024,1024),dtype =np.uint8))

    draw = ImageDraw.Draw(image_draw)

    draw.polygon((array1[1].tolist()),  outline= 1 , fill= 1 )

    return np.asarray(image_draw)
```

```python
final_image = []
train_txt = copy.deepcopy(train_txt2)
total_boat = len(patch_boat_type)-1
blank = np.zeros((1024,1024),dtype = np.uint8)
for i in progress_bar(range(0, len(train_mask))):
# for i in progress_bar(range(350, 360)):
    
    boat_add = 0
    mask = train_mask_1[i]
    newarray_image = copy.deepcopy(train_image[i])
    
    newarray_mask = copy.deepcopy(mask)
    
    total_boat_add = random.randint(5,12)
    
    otsu_image = copy.deepcopy(train_image[i])
    otsu_image = np.asarray(otsu_image,dtype = np.uint16)
    otsu_image = otsu_image[:,:,0] + otsu_image[:,:,1] + otsu_image[:,:,2]
    blur = cv2.GaussianBlur(otsu_image,(15,15),0)
    ret2,th2 = cv2.threshold(otsu_image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    ret2,th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    th2[th2!=0]=1
    th3[th3!=0]=1
    
    th2 = np.asarray(th2,dtype = np.uint8)
    th3 = np.asarray(th3,dtype = np.uint8)
    
    th2[th2!=0]=1
    th3[th3!=0]=1
    
    
    diff_th = np.sum(th2>0) -np.sum(th3>0) 
#     print(diff_th)
    if(diff_th> 9000 ):
        th2 = 0
        total_boat_add = 0
        
    mask = mask + th2
    mask[mask>1] = 1
    
    attemp = 0
    
    while( boat_add < total_boat_add):
        attemp += 1
        
        if(attemp>100):
            total_boat_add = 0
        newarray_mask = copy.deepcopy(mask)

        boat_id = random.randint(0,total_boat)
        
        patch_1 = patch_boat_1[boat_id]
        patch_0 = patch_boat_0[boat_id]
        patch_image = patch_boat_image[boat_id]
        patch_type = patch_boat_type[boat_id]

        
        zoom_scale = 0
        if(5000 >= np.sum(patch_1)):
            zoom_scale = random.random()/4 + 1
        
        if(10000 >= np.sum(patch_1) > 5000):
            zoom_scale = random.random()/2 + 0.75
            
        if( 20000 >=np.sum(patch_1) > 10000):
            zoom_scale = random.random()/4 + 0.45
            
        if( np.sum(patch_1) > 20000):
            zoom_scale = random.random()/4 + 0.25
        
        h,w  = (random.random()/4+0.875)*zoom_scale, (random.random()/4+0.875)*zoom_scale
        

        
        resize_x = int(h * patch_0.shape[1] )
        resize_y = int(w * patch_0.shape[0] )
        rotate_angle = random.randint(1,359)
        
        mask0_temp = cv2.resize(patch_0, (resize_x,resize_y))
        mask0_temp = np.array(Image.fromarray(mask0_temp).rotate(rotate_angle,expand=True,fillcolor =1))
        mask0_temp[mask0_temp!= 0] = 1
        
        mask1_temp = cv2.resize(patch_1, (resize_x,resize_y))
        mask1_temp = np.array(Image.fromarray(mask1_temp).rotate(rotate_angle,expand=True))
        mask1_temp[mask1_temp!= 0] = 1

        image_temp = cv2.resize(patch_image, (resize_x,resize_y))
        image_temp = np.array(Image.fromarray(image_temp).rotate(rotate_angle,expand=True))
        image_temp = np.asarray(image_temp,dtype=np.int16)
        image_temp = image_temp + (np.random.randint(-35,35,(image_temp.shape))) #noise level 35
        image_temp[image_temp>255] = 255
        image_temp[image_temp<0] = 0
        image_temp = image_temp * np.stack((mask1_temp,mask1_temp,mask1_temp), 2)
        image_temp = np.asarray(image_temp,dtype=np.uint8)
        
        area_patch = np.sum(mask1_temp>0)
        
        xnew = mask0_temp.shape[0]
        ynew = mask0_temp.shape[1]
        
        x,y  = random.randint(0,1024-xnew),random.randint(0,1024-ynew)
        
        

        
        newarray_mask[x:x+ xnew,y:y+ynew] = newarray_mask[x:x+ xnew,y:y+ynew] + mask1_temp

        areaall = np.sum(newarray_mask>1)
#         print(areaall)
        if( areaall == 0):
#             print('yes')
            temp_blank = copy.deepcopy(blank)
            temp_blank[x:x+ xnew,y:y+ynew] = temp_blank[x:x+ xnew,y:y+ynew] + mask1_temp
            newarray_mask[newarray_mask>1] = 1
            mask = newarray_mask
            newarray_image[x:x+ xnew,y:y+ynew] = newarray_image[x:x+ xnew,y:y+ynew] * np.stack((mask0_temp,mask0_temp,mask0_temp), 2)
            newarray_image[x:x+ xnew,y:y+ynew] = newarray_image[x:x+ xnew,y:y+ynew] + image_temp
            txt_data = np.concatenate(np.concatenate(cv2.findContours(temp_blank,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)[0][0]))
            train_txt[i].append([patch_type, txt_data])
            
            boat_add+=1
#     print(total_boat_add)
    final_image.append(newarray_image)
final_image = np.asarray(final_image, dtype = np.uint8)
```
